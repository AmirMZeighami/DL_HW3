```python
transform = transforms.Compose([transforms.Resize((224, 224)),
                                transforms.ToTensor(),
                               ])

tTransform=torchvision.transforms.Compose([lambda x:torch.LongTensor([x]),
                                           lambda x:torch.nn.functional.one_hot(x,10)])


trainDataset = torchvision.datasets.CIFAR10("./Cifar10", train = True, transform = transform, target_transform = tTransform, download = True)
testDataset = torchvision.datasets.CIFAR10("./Cifar10", train = False, transform = transform, target_transform = tTransform, download = True)
```

**جداسازی داده آموزش، ولیدیشن و تست**

```python
validationSplit = 0.2 
validationLength = int(validationSplit * len(trainDataset))
trainLength = len(trainDataset) - validationLength

print(f"Train Length : {trainLength}, ValidationLength : {validationLength}")
trainDataset, validationDataset = torch.utils.data.random_split(trainDataset,[trainLength, validationLength])
```

**این تابع صرفا پیش بینی ها و لیبل ها را دیافت می کند و تعداد داده های درست پیش بینی شده را بر میگرداند**

```python
def getCorrectPrediction(pred, y):
  return(sum(torch.argmax(pred, dim = 1) == torch.argmax(y, dim = 1)))
```

**این فانکشن دقت و خطای مدل با توجه به دیتاست داده شده به خود را محاسبه می کند**

```python
def eval(model, testDataset, batchSize = 100):
    positive = 0
    lossFn = torch.nn.CrossEntropyLoss()
    testDataLoader = torch.utils.data.DataLoader(testDataset, batch_size=batchSize, shuffle=True)
    losses = 0
    for i, data in enumerate(testDataLoader):
      Xt, yt = data
      Xt = Xt.to(device)
      yt = yt.to(device)
      yt = torch.reshape(yt, (yt.shape[0], 10))
      pred = model.to(device).forward(Xt)

      loss = lossFn(pred, yt.to(device).to(torch.float))
      losses = losses + loss.item()
      positive = positive + getCorrectPrediction(pred, yt)
    print(f"Test Loss : {losses / i}, Test Accuracy : {positive/len(testDataset)}")
    return losses/i, positive/len(testDataset)
```

**در این قسمت شبکه رزنت از پیش آموزش داده شده روی ایمیجنت خوانده شده و قسمت کانولوشنی آن غیر قابل آموزش شده و لایه دنس آن نیز به تعداد کلاس ها که برای این دیتاست 10 است تبدیل میشود**

```python
resnet50 = models.resnet50(pretrained = True)#, weights = 'IMAGENET1K_V2')
for param in resnet50.parameters():
  param.requires_grad = False

resnet50.fc = torch.nn.Linear(resnet50.fc.in_features, 10)
```

**فانکشن ترین بدون آموزگار، که داده را می گیرد، با استفاده از مباحث تمرین قبل شبکه آموزش داده می شود و سپس خطا و دقت برای داده های ترین و ولیدیشن محاسبه و چاپ می شود. در هر مرحله بهترین مدل با توجه به کمینه بودن خطای ولیدیشن ذخیره میشود**

```python
def train(model, trainDataLoader, validationDataLoader, lossFn, optim, scheduler, epochs, batchSize, bestModelPathDir):
  minValidationLoss = 10000
  for epoch in range(epochs):
      trainLossEpoch = 0
      validationLossEpoch = 0
      trainCorrect = 0
      validationCorrect = 0
      # randperm = torch.randperm(X.shape[0])  
      for i, data in (enumerate(trainDataLoader)):
        optim.zero_grad()
        X, y = data
        y = torch.reshape(y, (y.shape[0], 10)).to(torch.float)
        out = model.forward(X.to(device))
        loss = lossFn(out, y.to(device))
        # trainCorrect = trainCorrect + getCorrectPrediction(out, y.to(device))
        # trainLossEpoch = loss.item() + trainLossEpoch
        loss.backward()
        optim.step()
      # trainLossEpoch = trainLossEpoch / i
      for i, data in enumerate(trainDataLoader):
        X, y = data
        y = torch.reshape(y, (y.shape[0], 10)).to(torch.float)
        out = model.forward(X.to(device))
        loss = lossFn(out, y.to(device))
        trainCorrect = trainCorrect + getCorrectPrediction(out, y.to(device))
        trainLossEpoch = loss.item() + trainLossEpoch
      trainLossEpoch = trainLossEpoch / i
      for i, data in enumerate(validationDataLoader):
        X, y = data
        y = torch.reshape(y, (y.shape[0], 10)).to(torch.float)
        out = model.forward(X.to(device))
        loss = lossFn(out, y.to(device))
        validationCorrect = validationCorrect + getCorrectPrediction(out, y.to(device))
        validationLossEpoch = loss.item() + validationLossEpoch

      validationLossEpoch = validationLossEpoch / i
      if(validationLossEpoch < minValidationLoss):
        print(f"Validation Loss is improved by {minValidationLoss - validationLossEpoch}")
        minValidationLoss = validationLossEpoch
        torch.save(model, bestModelPathDir)
      print(f"Epoch {epoch} : TrainingLoss = {trainLossEpoch}, Training Categorical Accuracy : {trainCorrect / 40000}, validationLoss = {validationLossEpoch}, Validation Categorical Accuracy : {validationCorrect / 10000}")
  return model
```

**تابع هزینه با توجه به تابع هزینه تعریف شده در قسمت توضیحات سوال تعریف می شود توجه شود که رابطه ی داده شده در صورت سوال، با صورت نمایش داده شده در تابع قابل بازنویسی است**

```python
def KDLoss(alpha, temperature, predStudent, predTeacher, trueLabel, lossFn, softMax):
  loss = (1 - alpha) * lossFn(predStudent, trueLabel) + alpha*temperature*temperature*lossFn(predStudent/temperature, softMax(predTeacher/temperature))
  return loss
```

**Know Distillation Training**
<br > این تابع همانند تابع آموزش قبلی است با این تفاوت که یک بار داده های آموزش توسط مدل آموزگار پیش بینی شده و تابع هزینه برای رزنت18 به صورت فوق محاسبه می شود

```python
def KDTrain(teacher, student, trainDataLoader, validationDataLoader, lossFn, optim, scheduler, epochs, batchSize, alpha, temperature, bestModelPathDir):
  student.train()
  softMax = torch.nn.Softmax(dim=1)
  minValidationLoss = 10000
  for epoch in range(epochs):
      trainLossEpoch = 0
      validationLossEpoch = 0
      trainCorrect = 0
      validationCorrect = 0
      for i, data in enumerate(trainDataLoader):
        optim.zero_grad()
        X, y = data
        X = X.to(device)
        y = y.to(device)
        y = torch.reshape(y, (y.shape[0], 10)).to(torch.float)
        predTeacher = teacher.forward(X.to(device))
        predStudent = student.forward(X.to(device))
        loss = KDLoss(alpha, temperature, predStudent, predTeacher, y.to(device), lossFn, softMax)
        # trainCorrect = trainCorrect + getCorrectPrediction(out, y.to(device))
        # trainLossEpoch = loss.item() + trainLossEpoch
        loss.backward()
        optim.step()
      # trainLossEpoch = trainLossEpoch / i
      for i, data in enumerate(trainDataLoader):
        X, y = data
        y = torch.reshape(y, (y.shape[0], 10)).to(torch.float)
        out = student.forward(X.to(device))
        loss = lossFn(out, y.to(device))
        trainCorrect = trainCorrect + getCorrectPrediction(out, y.to(device))
        trainLossEpoch = loss.item() + trainLossEpoch
      trainLossEpoch = trainLossEpoch / i
      for i, data in enumerate(validationDataLoader):
        X, y = data
        y = torch.reshape(y, (y.shape[0], 10)).to(torch.float)
        out = student.forward(X.to(device))
        loss = lossFn(out, y.to(device))
        validationCorrect = validationCorrect + getCorrectPrediction(out, y.to(device))
        validationLossEpoch = loss.item() + validationLossEpoch
      validationLossEpoch = validationLossEpoch / i
      if(validationLossEpoch < minValidationLoss):
          print(f"Loss is improved by {minValidationLoss - validationLossEpoch}")
          minValidationLoss = validationLossEpoch
          torch.save(student, bestModelPathDir)
      print(f"Epoch {epoch} : TrainingLoss = {trainLossEpoch}, Training Categorical Accuracy : {trainCorrect / 40000}, validationLoss = {validationLossEpoch}, Validation Categorical Accuracy : {validationCorrect / 10000}")
  return student
```
